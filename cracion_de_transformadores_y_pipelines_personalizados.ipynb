# imports

```python
import arff
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
```

# funciones auxiliares

```python
def load_kdd_dataset(data_path):
    """Lectura del dataSet NSL-KDD."""
    with open(data_path, 'r') as train_set:
        dataset = arff.load(train_set)
    attributes = [attr[0] for attr in dataset["attributes"]]
    return pd.DataFrame(dataset["data"], columns = attributes)
```

```python
def train_val_test_split(df, rstate = 42, shuffle = True, stratify = None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size = 0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size = 0.5, random_state = rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

# lectura del dataset

```python
df = load_kdd_dataset("datasets/NSL-KDD/KDDTrain+.arff")
df
```

 # division del dataset

```python
train_set, val_set, test_set = train_val_test_split(df, stratify = 'protocol_type')
```

### 1.- construyendo transformadores personalizados 

la creaci√≥n de transformadores personalizados propios, permite mantener el codigo mucho mas limpio y estrucurado a la hora de preparar los datos para los algoritmos de ML. Adem√°s, facilitan la reutlizacion de codigo para otros proyectos.
antes de comenzar , esnecesario recuperar el dataset y separar las etiquetas del resto de los datos no necesasriamnete se quieren  aplicar las mismas transformaciones en ambos conjuntos 

```python
X_train = train_set.drop("class", axis = 1)
y_train = train_set["class"].copy()
```

```python
# para ilustrar esta secci√≥nenlos es necesario a√±adir algunos valores nulos a algunas caracter√≠sticas del dataset
X_train.loc[(X_train["src_bytes"]>400) & (X_train["src_bytes"]<800), "src_bytes"]= np.nan
X_train.loc[(X_train["dst_bytes"]>500) & (X_train["dst_bytes"]<2000), "dst_bytes"]= np.nan
X_train
```

#### transformadores para atributos numericos 

```python
# transformadores creado para eliminar las filas con valores nulos 
from sklearn.base import BaseEstimator, TransformerMixin

class DeleteNanRows(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        return X.dropna()
```

```python
delete_nan = DeleteNanRows()
X_train_prep = delete_nan.fit_transform(X_train)
X_train_prep
```

```python
# transformador dise√±ado para escalar de manera sencilla unicamente unas columnas seleccionadas 
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import RobustScaler
import pandas as pd

class CustomScaler(BaseEstimator, TransformerMixin):
    def __init__(self, attributes):
        self.attributes = attributes

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        scale_attrs = X_copy[self.attributes]
        robust_scaler = RobustScaler()  # Aseg√∫rate de que este est√© importado correctamente
        X_scaled = robust_scaler.fit_transform(scale_attrs)
        X_scaled = pd.DataFrame(X_scaled, columns=self.attributes, index=X_copy.index)
        
        for attr in self.attributes:
            X_copy[attr] = X_scaled[attr]
        
        return X_copy
```

```python
custom_scaler = CustomScaler(["src_bytes", "dst_bytes"])
X_train_prep = custom_scaler.fit_transform(X_train_prep)
X_train.head(10)
```

# 6.- construyendotelas pipelines personalizados 

las pipelines permiten agrupar en un flujo de ejecuci√≥n todas las operaciones de transformaci√≥n que se necesitan realizar sobre el DataSet, esto facilita muchisimo las transformaciones para diferentes conjutos . las cosas a tener en cuenta de estas estrucutura son:

* recibe un dataset de pares(nombre, estimador).
* todos menos el ultimo deben ser transformadores
* el pipeline expone los mismos metodos que el **ultimo estimador**.
* cuando se llama el metodo fit() del pipeline, se llama secuencialmente el metodo fit_transform() de los estimadores y se les pasa de manera secuencial el oupt del anterior como el input del siguiente. el ultimo invoca el metodo fit()

```python
# construccion de pipeline para los atributos numericos
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.impute import SimpleImputer
```

```python
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = "median")),
    ('rbst_scaler', RobustScaler())
])
```

```python
# la clase imputer no admite valores categ√≥ricos, se eliminan los atributos categ√≥ricos 
X_train_num = X_train.select_dtypes(exclude = ['object'])

X_train_prep = num_pipeline.fit_transform(X_train_num)
X_train_prep = pd.DataFrame(X_train_prep, columns=X_train_num.columns, index=X_train_num.index)
```

a continuacion se presenta el metodo ColumnTransformer, **que ejecuta todos los pipelines en paralelo y concatena el resultado** para ello el resultado de los pipelines debe ser un valor numerico 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(X_train.select_dtypes(exclude = ['object']))
cat_attribs = list(X_train.select_dtypes(include = ['object']))
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])
```

```python
X_train_prep = full_pipeline.fit_transform(X_train)
```

```python
X_train_prep = pd.DataFrame(X_train_prep, columns = list(pd.get_dummies(X_train)), index=X_train.index)
```
